```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('C:/Users/Admin/Desktop/Final Exam prep/urbanandruralpopulation new.csv')

# Display first few rows of the dataset
print(df)
```

```python
# Data cleaning: Ensuring there are no missing values or anomalies
df.dropna(inplace=True)
df.head()
```

```python
# Bar plot for Urban vs Rural population comparison
plt.figure(figsize=(10,6))
sns.barplot(x="Year", y="Urban population", data=df, color='skyblue', label='Urban')
sns.barplot(x="Year", y="Rural population", data=df, color='lightgreen', label='Rural')
plt.title('Urban vs Rural Population Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(loc='upper right')  # Manually specify the legend location
plt.xticks(rotation=45)
plt.show()
```

```python
# Line chart for Total Population Growth Over the Years
plt.figure(figsize=(10,6))
sns.lineplot(x="Year", y="Total Population", data=df, marker='o', color='purple')
plt.title('Total Population Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.xticks(rotation=45)
plt.show()
```
