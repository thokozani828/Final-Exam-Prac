```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller 

# Step 1: Load your dataset
data = pd.read_csv("C:/Users/Admin/Desktop/Final Exam prep/demand_forecasting_data.csv")
data.head()
```

```python
# Aggregate or sum the demand per date if you have multiple records per day
df_daily = data.groupby('Date').agg({'Demand': 'sum'}).reset_index()

# Print the first few rows of the aggregated dataset
df_daily.head()
```

```python
# Sort the data by date
df_daily = df_daily.sort_values('Date')

# Print the sorted data to verify the output
df_daily.head()
```

```python
# Visualize the time series data
plt.figure(figsize=(10, 6))
plt.plot(df_daily['Date'], df_daily['Demand'], label='Order Demand')
plt.xlabel('Date')
plt.ylabel('Order Demand')
plt.title('Order Demand Over Time')
plt.legend()
plt.show()
```

```python
# Check stationarity using Augmented Dickey-Fuller test
result = adfuller(df_daily['Demand'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Differencing to make the data stationary if necessary
if result[1] > 0.05:
    print("Data is non-stationary. Applying differencing.")
    df_daily['Demand_diff'] = df_daily['Demand'].diff()
    series_to_use = df_daily['Demand_diff'].dropna()  # Use the differenced series for ARIMA modeling
else:
    print("Data is stationary. No differencing needed.")
    series_to_use = df_daily['Demand']

# Define and fit ARIMA model
model = ARIMA(series_to_use, order=(5, 1, 0))  # Order: (p, d, q) - Adjust these values as needed
arima_result = model.fit()

# Print model summary
print(arima_result.summary())
```

```python
# Forecast future values
forecast_steps = 30  # Number of days to forecast
forecast = arima_result.forecast(steps=forecast_steps)

forecast.head()
```
